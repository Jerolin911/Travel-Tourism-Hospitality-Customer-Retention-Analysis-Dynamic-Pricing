```python
#LOADING THE DATASET 
import pandas as pd
import numpy as np

df = pd.read_csv("hotel_booking.csv")
print("Initial Shape:", df.shape)
print("\nFirst 5 rows:\n", df.head())
```

```python
#DISPLAYING DATASET AND CHECKING FOR MISSING VALUES
print("\nDataset Info:\n")
print(df.info())

print("\nMissing Values:\n")
print(df.isnull().sum())
```

```python
#DROPPING UNECESSARY COLUMNS
df.columns = df.columns.str.lower().str.replace(' ', '_')
cols_to_drop = ['name', 'email', 'phone-number', 'credit_card']
df.drop(columns=cols_to_drop, inplace=True, errors='ignore')
```

```python
#DROPPING DUPLICATE VALUES IF ANY 
df.drop_duplicates(inplace=True)

#HANDLING MISSING VALUES FROM THE DATASET INFO
df['children'] = df['children'].fillna(0)

df['agent'] = df['agent'].fillna(0)

df['company'] = df['company'].fillna(0)

df['country'] = df['country'].fillna('Unknown')
```

```python
#CLEANING COLUMNS OF CATEGORICAL DATA
df['hotel'] = df['hotel'].str.strip()
df['customer_type'] = df['customer_type'].str.strip()
df['market_segment'] = df['market_segment'].str.strip()
```

```python
#REMOVING RECORDS HAVING NO GUESTS

df = df[(df['adults'] + df['children'] + df['babies']) > 0]
df = df[df['lead_time'] >= 0]
```

```python
#HANDLING OUTLIERS
df = df[df['adr'] >= 0]
df = df[df['adr'] < 5000]
```

```python
#----- FEATURE ENGINEERING -----
```

```python
#CONVERTING MONTH NAMES TO NUMBERS FOR EASIER COMPARISON & PLOTTING OF DATA

df['arrival_date'] = pd.to_datetime(df['arrival_date_year'].astype(str) + '-' +df['arrival_date_month'].astype(str) + '-' +
    df['arrival_date_day_of_month'].astype(str),errors='coerce')


df['total_stay'] = df['stays_in_weekend_nights'] + df['stays_in_week_nights']
df['total_guests'] = df['adults'] + df['children'] + df['babies']
```

```python
print("\nFinal Shape:", df.shape)

print("\nRemaining Missing Values:\n")
print(df.isnull().sum())

print("\nCleaned Data:\n")
print(df.head())
```

```python
#SAVING THE CLEANED DATA IN ANOTHER CSV FILE
df.to_csv("cleaned_hotel_booking.csv", index=False)
print("Data Cleaning Completed")
```
